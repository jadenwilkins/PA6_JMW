```python
import math

class calc():
    def __init__(self,input1,input2):
        self.input1=input1
        self.input2=input2
    def addition(num1, num2):
        print(f'\nAddition= {num1+num2}')
    
    def subtraction(num1, num2):
        print(f'\nSubtraction= {num1-num2}')
    
    def multiplication(num1, num2):
        print(f'\nMultiplication= {num1*num2}')
        
    def division(num1, num2):
        print(f'\nDivision= {num1/num2}')
        
    def square(num1):
        print(f'\nSquare= {num1**2}')
        
    def remainder(num1, num2):
        print(f'\nRemainder= {num1 % num2}')
        
    def square_root(num1):
        print(f'\nSquare Root= {math.sqrt(num1)}')
        
    def power(num1, num2):
        print(f'\nPower= {num1**num2}')
        
    def LCM(num1, num2):
        print(f'\nLeast common multiple= {math.lcm(num1,num2)}')
        
    def average(num1, num2):
        print(f'\nAverage= {(num1+num2)/2}')
        
    def sine(num1):
        print(f'\nSine= {math.sin(num1)} in radians')
    
    def cos(num1):
        print(f'\nCosine= {math.cos(num1)} in radians')
        
    def tan(num1):
        print(f'\nTangent= {math.tan(num1)} in radians')
        
    def percentage(num1):
        print(f'\nPercentage= {num1*.01} %')
        
    def rect_peri(num1, num2):
        print(f'\nPerimeter of rectangle= {num1*2+num2*2}')
        

    print('Choose from the following menu: ')
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Square")
    print("6. Remainder")
    print("7. Square Root")
    print("8. Power")
    print("9. Least common multiple")
    print("10. Average")
    print("11. Sine")
    print("12. Cosine")
    print("13. Tangent")
    print("14. Percentage")
    print("15. Perimeter of Rectangle")
    print("16. End Program")

while True: 
    choice=int(input("Enter a number (1-16)"))
    
    
    if choice==1:
        num1=int(input("Enter 1st number:"))
        num2=int(input("Enter 2nd number:"))
        calc.addition (num1,num2)
    elif choice==2:
        num1=int(input("Enter 1st number:"))
        num2=int(input("Enter 2nd number:"))
        calc.subtraction (num1,num2)
    elif choice==3:
        num1=int(input("Enter 1st number:"))
        num2=int(input("Enter 2nd number:"))
        calc.multiplication (num1,num2)
    elif choice==4:
        num1=int(input("Enter 1st number:"))
        num2=int(input("Enter 2nd number:"))
        if num1==0:
            print("Infinity")
        else:
            calc.division (num1,num2)
    elif choice==5:
        num1=int(input("Enter 1st number:"))
        calc.square (num1)
    elif choice==6:
        num1=int(input("Enter 1st number:"))
        num2=int(input("Enter 2nd number:"))
        calc.remainder (num1,num2)
    elif choice==7:
        num1=int(input("Enter 1st number:"))
        calc.square_root (num1)
    elif choice==8:
        num1=int(input("Enter 1st number:"))
        num2=int(input("Enter 2nd number:"))
        calc.power (num1,num2)
    elif choice==9:
        num1=int(input("Enter 1st number:"))
        num2=int(input("Enter 2nd number:"))
        calc.LCM (num1,num2)
    elif choice==10:
        num1=int(input("Enter 1st number:"))
        num2=int(input("Enter 2nd number:"))
        calc.average (num1, num2)
    elif choice==11:
        num1=int(input("Enter 1st number:"))
        calc.sine (num1)
    elif choice==12:
        num1=int(input("Enter 1st number:"))
        calc.cos (num1)
    elif choice==13:
        num1=int(input("Enter 1st number:"))
        calc.tan (num1)
    elif choice==14:
        num1=int(input("Enter 1st number:"))
        calc.percentage (num1)
    elif choice==15:
        num1=int(input("Enter 1st number:"))
        num2=int(input("Enter 2nd number:"))
        calc.rect_peri (num1, num2)
    elif choice==16:
        break
        
    else:
        print("Invalid Input. Please try again.")
    
        
        
              
              
    
    
        
              
```
